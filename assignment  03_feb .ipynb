'''What is a generator function in python? Why yield keyword is used? Give an example of a generator
function.
ans- When we create this function using yield it does nott stores the complete data, instead it creates a generator so that when we want that data we can use that generator function and it will produce the required info when we apply it.

```python
def table(n):
    for i in range(1,n):
        yield i*n
```

```python
table(100)
```

```python
for i in table(10):
    print((i))
```

```python
num = int(input("Enter a number: "))
temp = num
reverse = 0

while temp > 0:
    digit = temp % 10
    reverse = reverse * 10 + digit
    temp = temp // 10

if num == reverse:
    print("The number is a palindrome")
else:
    print("The number is not a palindrome")
```
